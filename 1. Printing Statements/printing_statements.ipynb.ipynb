# **Printing Statements**

### **Example 1: Print a single value**

```python
print("Hello World!")
```

### **Example 2: Printing Variables**

```python
x = 10
y = 20
print("The value of x is: ", x)
print("The value of y is: ", y)
```

### **Example 3: Printing multiple variables using f-strings (Python 3.6+)**

```python
name = "Alex"
age = 30
print(f"My name is {name} and I am {age} years old")
```

### **Example 4: Using .format method for string formatting**

```python
name = "Alex"
age = 30
print("My name is {} and I am {} years old".format(name,age))
```

### **Example 5: Print using sep**

```python
print(12,24,-2, sep = ":")
```

### **Example 6: Print using endline**

```python
print("Hello\n")
```

### **Example 7: Print using end**

```python
print("Hello\n", end = ' ')
```
